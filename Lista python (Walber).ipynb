primos de 1 a 100. Escreva esses números primos em um arquivo chamado
primos.txt.**

```python
import numpy as np

def prime(x):
  primes = np.array([1, 2])
  n = primes[-1] 
  while n <= x:

    n += 1
    prime_test = n % primes[1:] != 0
    if prime_test.all():
      primes = np.append(primes, n)

  return primes

print(prime(100))

output = prime(100)
np.savetxt('primos.txt', output, delimiter=',')
file.close() 
```
