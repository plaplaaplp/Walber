As equações de transferência de calor, que determinam as temperaturas T1 e T2 em duas divisões de uma casa, são as seguintes:
Sabendo que: $H = \frac{K A(T_q - T_f)}{L}$

\begin{equation}
  \left\{ \begin{array}{c}\frac{dT_1}{dt} = 2 -0.5(T_1 - 8) - 0.5(T_1 - T_2)\\
  \frac{dT_2}{dt} = -0.1(T_2 - 8) - 0.5(T_2 - T_1) \end{array}\right.
  \end{equation}
Resolvendo até o momento em que $\frac{dT}{dt} = cst. = 0$(equilíbrio)

\begin{equation}
  \left\{ \begin{array}{c} -0.7T_1 + 0.5T_2 = -3.6\\
  0.5T_1 - 0.6T_2 = -0.8 \end{array}\right.
  \end{equation}
Na forma Ax=b
\begin{equation} 
\left(\begin{array}{cccccccccccccccccc}
-0.7&0.5\\
0.5&-0.6\\ 
\end{array}\right)*\left(\begin{array}{cccccccccccccccccc}
T_1\\
T_2 
\end{array}\right) = \left(\begin{array}{cccccccccccccccccc}
-3.6\\
-0.8 
\end{array}\right)
\end{equation}

```python
#METODO GAUSS-JACOBI
import numpy as np

A = np.array([[-0.7,0.5], [0.5,-0.6]])
b = np.array([[-3.6], [-0.8]])
dim = len(A)



for i in range(0,dim):
  a = 0
  for j in range(0,dim):
    if i==j: pass
    else:
      a = a + (A[i,j])
  if a > A[i,i]: 
    print("não satisfaz critério das linhas em",i)
    break
  else:
    print("satisfaz critério das linhas em",i)


for j in range(0,dim):
  q = 0
  for i in range(0,dim):
    if j==i: pass
    else:
      q = q + (A[i,j])
  if q > A[j,j]: 
    print("não satisfaz critério da coluna em",j)
    break
  else:
    print("satisfaz critério das colunas em",j)





x0 = np.zeros(dim)
for i in range(0,dim):
  for j in range(0,dim):
    if i==j:
      x0[i]=  b[i] / A[i,j]
print("x0 inicial=",x0)

xk = np.zeros(dim)

e=1e-6
d = 1
n = 0
N=100

while d>e:
  n=n+1
  if n==N: break
  for i in range(0,dim):
    somaj = 0
    for j in range(0,dim):
      if i==j:
        pass
      else:
        somaj = somaj + A[i,j]*x0[j]
    xk[i]=(1/A[i,i]*(b[i] - somaj))
    #print(xk)
  z = (xk - x0)
  dx = np.max(np.abs(xk))#|xk|
  dz = np.max(np.abs(z))#|xk - x(k-1)|
  d = (dz / dx)
  x0 = np.copy(xk)
print("n de iteração = ", n)
print(f'xk = {xk}')

```
