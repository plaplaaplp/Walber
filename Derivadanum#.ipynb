**1.** Escreva um programa para calcular a primeira e segunda derivada da função *cos(x)* em x = $\frac{π}{6}$.
Utilize as diferentes aproximações dadas em aula (fd, bd, cd, cd-5points) para a primeira derivada. Escreva os 
erros absolutos e relativos. Utilize h = 0.1, 0.01 e 0.001 .


---



```python
import numpy as np
import math 

def f(x): #função 
   return np.cos(x)
#Método derivada a direita (forward difference -fd):
def df(h):
  x = np.pi / 6
  return (f(x+h) - f(x))/h


#Método derivada a esquerda (backward difference -bd):
def bd(h):
  x = np.pi / 6
  return (f(x) - f(x - h))/h

#Método da diferença central (central difference - cd):
def cd(h):
  x = np.pi / 6
  return (f(x+h) - f(x-h))/(2*h)


#Método da diferença central (5 points, 5-point stencil):
def cd5(h):
  x = np.pi / 6
  return  ((1)/(12*h))*(f(x-2*h) - 8*(f(x-h)) + 8*(f(x+h)) - f(x+2*h))

def dd2(h):
  x = np.pi / 6
  return  ( f(x-h) - 2*f(x) + f(x+h) ) / (h**2)

x = np.pi / 6
h = [0.1, 0.01, 0.001]
ref = -np.sin(x)
ref2 = -np.cos(x)
print("Primeira Derivada Método-df[h=0.1]=", df(h[0]),"erro absoluto = ", np.abs( df(h[0]) - ref),"erro relativo =", np.abs( df(h[0]) - ref)/ np.abs(ref))
print("Primeira Derivada Método-df[h=0.01]=", df(h[1]),"erro absoluto = ", np.abs( df(h[1]) - ref ),"erro relativo =", np.abs( df(h[1]) - ref)/ np.abs(ref))
print("Primeira Derivada Método-df[h=0.001]", df(h[2]),"erro absoluto = ", np.abs( df(h[2]) - ref ),"erro relativo =", np.abs( df(h[2]) - ref)/ np.abs(ref))

print("Primeira Derivada Método-bd[h=0.1]=", bd(h[0]),"erro absoluto = ", np.abs( bd(h[0]) - ref),"erro relativo =", np.abs( bd(h[0]) - ref)/ np.abs(ref))
print("Primeira Derivada Método-bd[h=0.01]=", bd(h[1]),"erro absoluto = ", np.abs( bd(h[1]) - ref ),"erro relativo =", np.abs( bd(h[1]) - ref)/ np.abs(ref))
print("Primeira Derivada Método-bd[h=0.001]", bd(h[2]),"erro absoluto = ", np.abs( bd(h[2]) - ref ),"erro relativo =", np.abs( bd(h[2]) - ref)/ np.abs(ref))

print("Primeira Derivada Método-cd[h=0.1]=", cd(h[0]),"erro absoluto = ", np.abs( cd(h[0]) - ref),"erro relativo =", np.abs( cd(h[0]) - ref)/ np.abs(ref))
print("Primeira Derivada Método-cd[h=0.01]=", cd(h[1]),"erro absoluto = ", np.abs( cd(h[1]) - ref ),"erro relativo =", np.abs( cd(h[1]) - ref)/ np.abs(ref))
print("Primeira Derivada Método-cd[h=0.001]", cd(h[2]),"erro absoluto = ", np.abs( cd(h[2]) - ref ),"erro relativo =", np.abs( cd(h[2]) - ref)/ np.abs(ref))

print("Primeira Derivada Método-cd5[h=0.1]=", cd5(h[0]),"erro absoluto = ", np.abs( cd5(h[0]) - ref),"erro relativo =", np.abs( cd5(h[0]) - ref)/ np.abs(ref))
print("Primeira Derivada Método-cd5[h=0.01]=", cd5(h[1]),"erro absoluto = ", np.abs( cd5(h[1]) - ref ),"erro relativo =", np.abs( cd5(h[1]) - ref)/ np.abs(ref))
print("Primeira Derivada Método-cd5[h=0.001]", cd5(h[2]),"erro absoluto = ", np.abs( cd5(h[2]) - ref ),"erro relativo =", np.abs( cd5(h[2]) - ref)/ np.abs(ref))

print("Segunda Derivada[h=0.1]=", dd2(h[0]),"erro absoluto = ", np.abs( dd2(h[0]) - ref2),"erro relativo =", np.abs( dd2(h[0]) - ref2)/ np.abs(ref2))
print("Segunda Derivada[h=0.01]=", dd2(h[1]),"erro absoluto = ", np.abs( dd2(h[1]) - ref2 ),"erro relativo =", np.abs( dd2(h[1]) - ref2)/ np.abs(ref2))
print("Segunda Derivada[h=0.001]", dd2(h[2]),"erro absoluto = ", np.abs( dd2(h[2]) - ref2 ),"erro relativo =", np.abs( dd2(h[2]) - ref2)/ np.abs(ref2))

```

**2. Seja uma partícula de *massa = 10 Kg*, que possui suas coordenadas em função do tempo dadas por:**

*   $x(t) = 30tcos(60)$
*   $y(t) = 30tsin(60) - \frac{gt^2}{2}$


Escreva um programa que calcula as forças sobre a partícula nas direções x e y


---




```python
import numpy as np

m = 10

def x(t): 
  return 30*t*np.cos(np.radians(60))

def y(t):
  g = 9.8 
  return 30*t*np.sin(np.radians(60)) - (g*(t**2))/2

def ddx(t):
  h = 1e-10
  return  ( x(t-h) - 2*x(t) + x(t+h) ) / (h**2)

def ddy(t):
  h = 1e-2
  return  ( y(t-h) - 2*y(t) + y(t+h) ) / (h**2)


for i in np.linspace(0, 1, 10):
  print("Força y(t) em ","t(s)", i, "Fy=", m*ddy(i),"N")
  print("Força x(t) em ","t(s)", i, "Fx=", m*ddx(i), "N")
```
