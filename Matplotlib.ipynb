**1. Escreva um programa que plote na mesma janela, gráficos das funções $cos(x), e^{-x²}sen(x), \sqrt{x}e^{\frac{-
x^2}{10.0}}$.**


```python
import numpy as np
import matplotlib.pyplot as plt
x = np.linspace(0,10,500) #cria vetor de 0 até 100 com 500 pontos do plano
cosx = np.cos(x) 
sq = np.sqrt(x)*np.exp(-x**2/10)
e = np.exp(-x**2)*np.sin(x)
fig = plt.figure(figsize=(8,8)) #cria figura com o plot
fig, ax = plt.subplots() # chama ax pra figura 
ax.plot(x, sq, color="cyan", label = "sqrt") # plota o ax no fig
ax.plot(x, cosx, color="magenta", label ="cos(x)") # plota a segunda função 
ax.plot(x, e, color = "yellow", label= "e")#plota a terceira função
ax.legend()#cria legenda definida no 'label' do ax.plot
plt.grid()#cria uma descretização
plt.show()#show figura
```

**2. Escreva um programa que plot em 2D (via colormap) a função $\sqrt{x^2+y^2}e^{-x^2 – y^2}$. Salve a 
imagem em um arquivo exercicio2.png.**

```python
import matplotlib as mpl
import numpy as np
x = y = np.linspace(-5,5,900)
X, Y = np.meshgrid(x,y)
Z = np.sqrt((X**2)+(Y**2))*np.exp(-(X**2)-(Y**2))
fig, ax = plt.subplots(figsize=(5,5))
norm = mpl.colors.Normalize(-abs(Z).max(), abs(Z).max()) #Máximos e mínimos do eixo Z configurada pra uma escola de cores
p = ax.imshow(Z, norm=norm, cmap=mpl.cm.bwr) 
fig.colorbar(p) #cria a barra de cores
plt.savefig('exercicio2.png', format='png') #salva a figura
plt.show()
```

**3. Escreva programas que plotem os campos vetoriais:**

\begin{equation} \vec{F}(x,y,z)=3x^2y\hat{i}+xy\hat{j}+z\hat{k} \end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d #importa a biblioteca para plotagem 3D
fig = plt.figure(figsize=(8,6)) #cria figura
ax= fig.gca(projection="3d") #seta a projeção para 3D
x, y, z = np.meshgrid(np.arange(-1, 1, 0.2), np.arange(-1, 1, 0.2), np.arange(-1, 1, 0.8),)#Descretiza o espaço de -1 a 1 pulando de 0.2 em 0.2 pontos
u=3*(x**2)*(y) #Atribui a um vetor (u) a coordenada i da função 
v=y*x # atribui a um vetor (v) a coordenada j da função vetorial
w=z
ax.quiver(x, y, z, u, v, w, length=0.15) #quiver a xyz em uvw com tamanho 0.15
plt.show()
```

\begin{equation} \vec{g}=\frac{\vec{r}}{r^3} \end{equation}

```python
import matplotlib.pyplot as plt
import numpy as np
from mpl_toolkits.mplot3d import axes3d
fig = plt.figure(figsize=(8,6))
ax= fig.gca(projection="3d")
x, y, z = np.meshgrid(np.arange(-1, 1, 0.4), np.arange(-1, 1, 0.4), np.arange(-1, 1, 0.4),)
u=(x)/(np.sqrt(x**2+y**2+z**2))**3
v=(y)/(np.sqrt(x**2+y**2+z**2))**3
w=(z)/(np.sqrt(x**2+y**2+z**2))**3
ax.quiver(x, y, z, u, v, w, length=0.05)
plt.show()
```
