```python
#interpolação polinomial

import numpy as np
import matplotlib.pyplot as plt

x = np.array([-1, 0, 3])
y = np.array([15, 8, -1])#fx

dim = len(x)



A = np.zeros((dim,dim), dtype=np.float64)
A[0:dim, 0] = 1



for j in range(1, dim):
  for i in range(0, dim):
    A[i, j] = x[i]**j

print("Matriz A = ")
print(A)




alpha = np.linalg.solve(A, y)
print("coeficientes a = ",alpha)


plt.plot(x, y ,"ro")
data = np.linspace(min(x),max(x), 100)
a = alpha[::-1].copy()
p = np.poly1d(a)
plt.plot(data , p(data), color = "blue")
plt.grid()
plt.show()

print(p)
```

```python
#interporalção polinomial Newton grau 4

import numpy as np
import matplotlib.pyplot as plt
import sympy as sympy


x = np.array([-1, 0, 1, 2, 3])
f = np.array([1, 1, 0, -1, -2])#fx

dim = len(x)

A = np.zeros((dim-1,dim-1), dtype=np.float64)

j = 0
for i in range(dim-1):
  A[0, i] = (f[i+1] - f[i]) / (x[i+1] - x[i])


i=0
for n in range(dim-2, -1, -1):
  j=j+1
  for i in range(n):
    j=j
    #print("n=", n)
    #print('J=', j)
    #print("i=", i)
    k=j+1
    A[j, i] = (A[j-1, i+1] - A[j-1, i]) / (x[i+(k)] - x[i])
    
#print("", A)

v = np.zeros(0)

v = np.append(v, f[0])

for i in range(dim-1):
  v = np.append(v, A[i,0])
print("diferencas divididas", v)



g = np.array([1, -0.75000001, -0.45833333, 0.25000001, -0.04166667])#vetor coeficientes polinomio



plt.plot(x, f ,"ro")
data = np.linspace(min(x),max(x), 100)
a = g[::-1].copy()
p = np.poly1d(a)
plt.plot(data , p(data), color = "blue")
plt.grid()
plt.show()



print('p',p)


sympy.init_printing()

u = sympy.Symbol('u')

px = v[0]*(1) + v[1]*(u-x[0]) + v[2]*(u-x[0])*(u-x[1]) + v[3]*(u-x[0])*(u-x[1])*(u-x[2]) + v[4]*(u-x[0])*(u-x[1])*(u-x[2])*(u-x[3])#polinomio grau 4 ........ p(5) => v[5]*(u-x[0])*(u-x[1])*(u-x[2])*(u-x[3])*(u-x[4])

sympy.simplify(px)
```
