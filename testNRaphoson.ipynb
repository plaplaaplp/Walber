```python
import math
import numpy as np
import matplotlib.pyplot as plt
class ddx:
    def __init__(self, f, h=1e-6):
        self.f = f
        self.h = float(h)

    def __call__(self, x):
        f = self.f
        h = self.h
        return (f(x+h) - f(x))/h


def g(t): return (math.tan(1.5*t)-t)
ddg = ddx(g)
t = 1
def gn(t): return t - (g(t)/ddg(t))

x11 = gn(t)
x22 = gn(x11)
x33 = gn(x22)
x44 = gn(x33)
```
