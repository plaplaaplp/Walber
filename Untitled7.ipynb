A Física computacional é um ramo que tem ganhado destaque devido a resolução e simulação de problemas complexos ou problemas de difícil acesso, podemos citar como exemplo: simulações astronômicas e física estatística.
Logo, a física computacional se relaciona com várias áreas do conhecimento fazendo sua parte colaborativa em simulações cada vez melhores e próximas ao que poderíamos considerar ''real''.


```python
import sympy # importa o sympy
sympy.init_printing() #começa a printar a eq.
t, g, v, v0, theta = sympy.symbols("t, g, v, v0, theta") # criar simbolos/variáveis da equação 
x = sympy.Function('x')(t) # coloca x em função de (t)
x = v0*sympy.cos(theta) # define uma equação
y = v0*sympy.sin(theta)*2 -g*t
sympy.integrate(x, t), sympy.integrate(y, t), sympy.solve(y, t) # diferencia x em relação a (t)
```

```python
import numpy as np
import matplotlib.pyplot as plt

v0 = 100
theta = 45
g = 9.8


t = np.linspace(0,14.5,100)

xt = (v0)*(np.cos(np.radians(theta)))*(t)

yt = v0*(np.sin(np.radians(theta)))*(t) - (1/2)*g*((t)**2)


fig = plt.figure(figsize=(8,8)) #cria figura com o plot
fig, ax = plt.subplots() # chama ax pra figura 
ax.plot(t, xt, color="cyan", label = "x(t)") # plota o ax no 
ax.plot(t, yt, color="magenta", label ="y(t)") 
ax.legend()#cria legenda definida no 'label' do ax.plot
plt.grid()#cria uma descretização
plt.show()#show figura

#ax.plot(x, cosx, color="magenta", label ="cos(x)")
#print(t)

```
