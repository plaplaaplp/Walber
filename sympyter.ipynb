**1. Seja a equação da posição de uma partícula em movimento:**

```python
import sympy # importa o sympy
sympy.init_printing() #começa a printar a eq.
t, g, v, x0 = sympy.symbols("t, g, v, x0") # criar simbolos/variáveis da equação 
x = sympy.Function('x')(t) # coloca x em função de (t)
x = x0 + v*t - g*t**2/(2) # define uma equação
sympy.diff(x, t) # diferencia x em relação a (t)
```

**Velocidade:**

```python
sympy.diff(x, t) # para derivada de primeira ordem
```

**Aceleração:**

```python
sympy.diff(x, t, t) # ára derivada de segunda ordem
```
